```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
imahe = cv2.imread('/content/drive/MyDrive/Image/orange.png')
```

```python
cv2_imshow(imahe)
```

```python
#convert to grayscale
imahe_graysc = cv2.cvtColor(imahe, cv2.COLOR_BGR2GRAY)

#display
cv2_imshow(imahe_graysc)
```

```python
#resize image
imahe_rs = cv2.resize(imahe, (250, 250))

#display
cv2_imshow(imahe_rs)
```

```python
#flip an image
imahe_fl = cv2.flip(imahe, -1)

#display
cv2_imshow(imahe_fl)
```
